```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("nb_tweet_h_ouighours.csv")
df["date"] = pd.to_datetime(df["date"], format = "%Y-%m-%dT%H")
fig = px.line(df, x="date", y="value")
fig.show()
```

```python
df['roll_day'] = df["value"].rolling(3 * 3, center=True).mean()
fig = px.line(df, x="date", y="roll_day")
fig.show()
```

```python
df_1 = pd.read_csv("nb_tweet_h_uyghurs.csv")
df_1["date"] = pd.to_datetime(df_1["date"], format = "%Y-%m-%dT%H")
fig = px.line(df_1, x="date", y="value")
fig.show()
```

```python
df_1['roll_1h'] = df_1["value"].rolling(3, center=True).mean()
df_1['roll_3h'] = df_1["value"].rolling(9, center=True).mean()
df_1['roll_12h'] = df_1["value"].rolling(3 * 12, center=True).mean()
df_1['roll_day'] = df_1["value"].rolling(3 * 24, center=True).mean()
```

```python
fig = px.line(df_1, x="date", y="roll_3h")
fig.show()
```

```python
fig = px.line(df_1, x="date", y="roll_12h")
fig.show()
```

```python
fig = px.line(df_1, x="date", y="roll_day")
fig.show()
```

```python
df_1['saison'] = [i-j for i,j in zip(df_1['roll_day'],df_1['value'])]
fig = px.line(df_1, x="date", y="saison")
fig.show()
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf, pacf, acf
plot_acf(df_1['saison'][3 * 12:-3 * 12])
plt.show()
```
